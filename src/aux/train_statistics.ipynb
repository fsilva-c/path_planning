```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_statistics(file, time_steps, mean_reward):
    df = pd.read_csv(file)
    time_steps = df[time_steps]
    mean_reward = df[mean_reward]
    plt.plot(time_steps.to_numpy(), mean_reward.to_numpy())
    plt.show()
```

```python
plot_statistics(
    '../DQN_fsppenv_log/progress.csv',
    time_steps='time/total_timesteps',
    mean_reward='rollout/ep_rew_mean'
)
```
